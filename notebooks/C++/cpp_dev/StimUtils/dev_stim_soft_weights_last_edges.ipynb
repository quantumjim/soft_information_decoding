```python
from soft_info import RepetitionCodeStimCircuit
import cpp_soft_info

D = 5
T = 6
LOGICAL = '0'
XBASIS = False
RESETS = False
# NOISE_LIST = [3e-2, 1e-4, 1e-2, 1e-2, 2e-2, 1e-2, 1e-2]
NOISE_LIST = [0, 0, 0, 0, 0, 0, 0]
SUBSAMPLING = False

code = RepetitionCodeStimCircuit(D, T, XBASIS, RESETS, NOISE_LIST, SUBSAMPLING)
circ = code.circuits[LOGICAL]
```

```python
from soft_info import draw_matching_graph
import pymatching

model = circ.detector_error_model()
matching = pymatching.Matching(model)

draw_matching_graph(matching, D, T, figsize=(2, 2), scale_factor=0.35)
```
