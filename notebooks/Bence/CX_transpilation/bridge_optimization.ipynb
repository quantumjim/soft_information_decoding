```python
from qiskit import transpile, QuantumCircuit

qc = QuantumCircuit(4)

qc.swap(0, 2)
qc.cx(2, 3)
qc.cx(1, 2)
qc.swap(0, 2)
# qc.barrier()

qc.draw(fold=-1)
```

```python
from qiskit_ibm_provider import IBMProvider

provider = IBMProvider()
backend = provider.get_backend('ibm_torino')
```

```python
transpiled_qc = transpile(qc, backend=backend, optimization_level=1)
```

```python
transpiled_qc.draw(fold=-1, idle_wires=False)
```
