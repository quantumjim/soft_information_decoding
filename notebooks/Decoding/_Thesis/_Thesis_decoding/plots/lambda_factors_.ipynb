# Plotting

```python
import numpy as np

distances = []
errors = []
shots = []
labels = []
for idx, (method, data) in enumerate(err_per_method.items()):
    if idx not in [0, 1, 2]:
        continue
    labels.append(method)
    errs = np.array(data['errs'])
    distances.append(np.array(data['d']))
    errors.append(np.array(data['errs']))
    shots.append(np.array(data['tot_shots']))

plot_error_rates(distances, errors, shots, labels=labels, plot_e_L=True, T=ROUNDS, mark_outliers=True)
```
