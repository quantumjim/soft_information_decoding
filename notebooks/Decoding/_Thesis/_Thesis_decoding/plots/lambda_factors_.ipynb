# Load the data

```python
%reload_ext autoreload
%autoreload 2

from dev.data import get_err_dicts
from dev.plotter import plot_error_rates


DEVICE = 'ibm_sherbrooke'
STATE = 'X0'
ROUNDS = 50

file_name = f'../results/{DEVICE}_{STATE}_{ROUNDS}.json'

err_per_job, err_per_method = get_err_dicts(file_name)
print(err_per_method)
```

# Plotting

```python
import numpy as np

distances = []
errors = []
shots = []
labels = []
for idx, (method, data) in enumerate(err_per_method.items()):
    if idx not in [3, 4]:
        continue
    labels.append(method)
    errs = np.array(data['errs'])
    distances.append(np.array(data['d']))
    errors.append(np.array(data['errs']))
    shots.append(np.array(data['tot_shots']))

plot_error_rates(distances, errors, shots, labels=labels, plot_e_L=True, T=ROUNDS, ignIdx=0)
```
