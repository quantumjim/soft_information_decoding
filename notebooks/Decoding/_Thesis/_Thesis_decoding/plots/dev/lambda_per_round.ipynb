```python
%reload_ext autoreload
%autoreload 2

from data import get_err_dicts
from plotter import logical_err_rate_per_round, lambda_func


DEVICE = 'ibm_sherbrooke'

distances = [51, 49, 47, 45, 43, 41, 39, 37, 35, 33, 31, 29, 27, 25, 23, 21, 19, 17, 15, 13, 11, 9, 7, 5, 3]
```

```python
import numpy as np
from scipy.optimize import curve_fit

outlier_nbs = 5


STATES = ['X0', 'X1', 'Z0', 'Z1']
# STATES = ['X0']

lambda_per_state = {}
for STATE in STATES:
                                        
    round_err_dict = {}
    for ROUNDS in [10, 20, 30, 40, 50, 75, 100]:
        file_name = f'../../results/{DEVICE}_{STATE}_{ROUNDS}.json'
        try:
            _, err_per_method = get_err_dicts(file_name)
            round_err_dict[ROUNDS] = err_per_method
        except:
            continue


    lambda_per_round = {T: {} for T in round_err_dict.keys()}
    for method in ['s_KPS', 'h_KPS', 'h_K_meanPS']:
        for T in round_err_dict.keys():
            ds = np.array(round_err_dict[T][method]['d'])
            errs = np.array(round_err_dict[T][method]['errs'])
            shots = np.array(round_err_dict[T][method]['tot_shots'])
            ps = errs/shots
            e_Ls = logical_err_rate_per_round(ps, T)


            # outliers
            valid_indices = np.where(errs > outlier_nbs)
            d_fit = ds[valid_indices]
            e_Ls = e_Ls[valid_indices]

            # Lambda fit
            log_eLs_fit = np.log10(e_Ls)
            popt, pcov = curve_fit(lambda_func, d_fit, log_eLs_fit)
            log10_C, neg_half_log10_Lambda = popt
            C = 10 ** log10_C
            Lambda = 10 ** (-2 * neg_half_log10_Lambda)  # since neg_half_log10_Lambda = -0.5 * log10(Lambda)
            error_neg_half_log10_Lambda = np.sqrt(pcov[1, 1])
            lambda_error = abs(-2 * np.log(10) * Lambda * error_neg_half_log10_Lambda) # error propagation

            lambda_per_round[T][method] = Lambda

            fit_distances = np.linspace(min(distances), max(distances), 100)
            fit_ps = lambda_func(fit_distances, log10_C, neg_half_log10_Lambda)

    lambda_per_state[STATE] = lambda_per_round

lambda_per_state
```

```python
import matplotlib.pyplot as plt

for state in STATES:
    plt.figure(dpi=100)
    rounds = np.array(list(lambda_per_state[state].keys()))
    lambdas_S = np.array([lambda_per_state[state][T]['s_KPS'] for T in rounds])
    lambdas_H = np.array([lambda_per_state[state][T]['h_KPS'] for T in rounds])
    lambdas_H_mean = np.array([lambda_per_state[state][T]['h_K_meanPS'] for T in rounds])
    ratio_S = lambdas_S/lambdas_H
    ratio_H_mean = lambdas_H_mean/lambdas_H

    plt.plot(rounds, ratio_S-1, label='s_KPS/h_KPS')
    plt.plot(rounds, ratio_H_mean-1, label='h_K_meanPS/h_KPS')

    plt.title(f'{state} logical error rate ratio')
    plt.ylabel('%')
    plt.grid(which='both', linestyle='--', linewidth=0.5)
    plt.legend()
```

```python
# Mean over all states

import numpy as np
import matplotlib.pyplot as plt

# Assuming 'STATES' and 'lambda_per_state' are already defined

# Initialize dictionaries to store ratios for each round across all states
ratio_S_per_round = {}
ratio_H_mean_per_round = {}

# Collect ratios
for state in STATES:
    rounds = np.array(list(lambda_per_state[state].keys()))
    lambdas_S = np.array([lambda_per_state[state][T]['s_KPS'] for T in rounds])
    lambdas_H = np.array([lambda_per_state[state][T]['h_KPS'] for T in rounds])
    lambdas_H_mean = np.array([lambda_per_state[state][T]['h_K_meanPS'] for T in rounds])
    
    ratio_S = lambdas_S / lambdas_H
    ratio_H_mean = lambdas_H_mean / lambdas_H

    for round_idx, round_val in enumerate(rounds):
        if round_val not in ratio_S_per_round:
            ratio_S_per_round[round_val] = []
            ratio_H_mean_per_round[round_val] = []
        
        ratio_S_per_round[round_val].append(ratio_S[round_idx])
        ratio_H_mean_per_round[round_val].append(ratio_H_mean[round_idx])

# Compute mean ratios
mean_ratio_S = np.array([np.mean(ratio_S_per_round[round_val]) for round_val in sorted(ratio_S_per_round)])
mean_ratio_H_mean = np.array([np.mean(ratio_H_mean_per_round[round_val]) for round_val in sorted(ratio_H_mean_per_round)])
sorted_rounds = np.array(sorted(ratio_S_per_round.keys()))

# Plotting
plt.figure(dpi=100)
plt.plot(sorted_rounds, mean_ratio_S - 1, label='Mean s_KPS/h_KPS')
plt.plot(sorted_rounds, mean_ratio_H_mean - 1, label='Mean h_K_meanPS/h_KPS')

plt.title('Mean Logical Error Rate Ratio Across States')
plt.ylabel('Mean Ratio - 1 (%)')
plt.xlabel('Rounds')
plt.grid(which='both', linestyle='--', linewidth=0.5)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Initialize dictionaries to store normalized ratios for each round across all states
normalized_ratio_S_per_round = {}
normalized_ratio_H_mean_per_round = {}

# Collect and normalize ratios
for state in STATES:
    rounds = np.array(list(lambda_per_state[state].keys()))
    lambdas_S = np.array([lambda_per_state[state][T]['s_KPS'] for T in rounds])
    lambdas_H = np.array([lambda_per_state[state][T]['h_KPS'] for T in rounds])
    lambdas_H_mean = np.array([lambda_per_state[state][T]['h_K_meanPS'] for T in rounds])
    
    ratio_S = lambdas_S / lambdas_H
    ratio_H_mean = lambdas_H_mean / lambdas_H

    # Normalize ratios within each state
    norm_ratio_S = (ratio_S - np.mean(ratio_S)) / np.std(ratio_S)
    norm_ratio_H_mean = (ratio_H_mean - np.mean(ratio_H_mean)) / np.std(ratio_H_mean)

    for round_idx, round_val in enumerate(rounds):
        if round_val not in normalized_ratio_S_per_round:
            normalized_ratio_S_per_round[round_val] = []
            normalized_ratio_H_mean_per_round[round_val] = []
        
        normalized_ratio_S_per_round[round_val].append(norm_ratio_S[round_idx])
        normalized_ratio_H_mean_per_round[round_val].append(norm_ratio_H_mean[round_idx])

# Compute mean of normalized ratios
mean_normalized_ratio_S = np.array([np.mean(normalized_ratio_S_per_round[round_val]) for round_val in sorted(normalized_ratio_S_per_round)])
mean_normalized_ratio_H_mean = np.array([np.mean(normalized_ratio_H_mean_per_round[round_val]) for round_val in sorted(normalized_ratio_H_mean_per_round)])
sorted_rounds = np.array(sorted(normalized_ratio_S_per_round.keys()))

# Plotting
plt.figure(dpi=100)
plt.plot(sorted_rounds, mean_normalized_ratio_S, label='Mean Normalized s_KPS/h_KPS')
plt.plot(sorted_rounds, mean_normalized_ratio_H_mean, label='Mean Normalized h_K_meanPS/h_KPS')

plt.title('Mean Normalized Logical Error Rate Ratio Across States')
plt.ylabel('Normalized Mean Ratio')
plt.xlabel('Rounds')
plt.grid(which='both', linestyle='--', linewidth=0.5)
plt.legend()
plt.show()
```
