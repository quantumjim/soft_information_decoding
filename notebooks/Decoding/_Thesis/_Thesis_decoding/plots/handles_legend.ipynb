```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import matplotlib.lines as mlines
from soft_info import apply_formatting, save_plot

# Define the custom handles for the legend
apply_formatting()
markersize = 20
linewidth = 3
plus_state_handle = mlines.Line2D([], [], color='grey', marker='o', markersize=markersize, label=r'$|+\rangle$ state', linestyle='-', linewidth=linewidth)
minus_state_handle = mlines.Line2D([], [], color='grey', marker='d', markersize=markersize, label=r'$|-\rangle$ state', linestyle='-', linewidth=linewidth)

outlier_handle = mlines.Line2D([], [], color='grey', linestyle=':', marker='o',
                               markersize=markersize, markerfacecolor='none', markeredgewidth=1,
                               label='Outliers $n_{errs} \leq 5$', linewidth=2)  

ci_patch = mpatches.Patch(color='black', alpha=0.1, label='68\% CI (Wilson)')

x_basis_handle = mpatches.Patch(color='red', label='X basis')
z_basis_handle = mpatches.Patch(color='blue', label='Z basis')

# Gather all handles
handles = [ x_basis_handle, z_basis_handle, plus_state_handle, minus_state_handle, outlier_handle, ci_patch]

# Create the figure and axes
fig, ax = plt.subplots()
ax.axis('off')  # Turn off axis

# Display the legend
legend = ax.legend(handles=handles, loc='center', fontsize='28')
save_plot(fig, '4_RepetitionCodes/legend_labels.pdf', format='pdf')
```
