```python
from result_saver import SaverProvider

provider = SaverProvider()
```

```python
from Scratch import load_calibration_memory

DEVICE = 'ibm_torino'
DATE = '2024-03-24'

all_memories, gmm_dict, msmt_err_dict = load_calibration_memory(provider, tobecalib_backend=DEVICE, other_date=DATE,
                                                                post_process=True, double_msmt=True)
```

```python
import numpy as np
p_hard_0 = []
p_hard_1 = []
p_soft_0 = []
p_soft_1 = []
for qubit, data in msmt_err_dict.items():
    p_hard_0.append(data['p_hard_0'])
    p_hard_1.append(data['p_hard_1'])
    p_soft_0.append(data['p_soft_0'])
    p_soft_1.append(data['p_soft_1'])

p_hard_0 = np.mean(p_hard_0)
p_hard_1 = np.mean(p_hard_1)
p_soft_0 = np.mean(p_soft_0)
p_soft_1 = np.mean(p_soft_1)

print(f"DEVICE: {DEVICE}, DATE: {DATE}")
print(f"p_hard_0: {p_hard_0}")
print(f"p_hard_1: {p_hard_1}")
print(f"p_soft_0: {p_soft_0}")
print(f"p_soft_1: {p_soft_1}")

msmt_err_dict_mean = msmt_err_dict.copy()
for qubit, data in msmt_err_dict_mean.items():
    data['p_hard_0'] = p_hard_0
    data['p_hard_1'] = p_hard_1
    data['p_soft_0'] = p_soft_0
    data['p_soft_1'] = p_soft_1

print(msmt_err_dict_mean)
```
