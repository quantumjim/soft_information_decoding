```python
%reload_ext autoreload
%autoreload 2

import pymatching
from soft_info import draw_matching_graph, RepetitionCodeStimCircuit, save_plot

d = 7
T = 3

code = RepetitionCodeStimCircuit(d, T, noise_list= [1e-2] + [1e-2]*6, resets=False)
circ = code.circuits['0']

model = circ.detector_error_model()
matching = pymatching.Matching.from_detector_error_model(model)
syndromes = [0]*((d-1)*(T+1))
indices = [12, 8, 20, 3, 9, 15, 22, 16, 17]
for i in indices:
    syndromes[i] = 1
fig = draw_matching_graph(matching, d=d, T=T, scale_factor=0.5, dpi=1000, 
                          matched_edges=[(8, 20), (12, -1), (15, 22), (16, 17), (3, 9)],
                          syndromes=syndromes)
# save_plot(fig, '4_RepetitionCodes/decoding_graph_RepCode.pdf', format='pdf')
```
