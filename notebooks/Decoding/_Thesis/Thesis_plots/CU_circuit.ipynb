```python
from qiskit import QuantumCircuit

# Create the quantum circuit
circ = QuantumCircuit(3, 1)
circ.h(0)              # Apply Hadamard gate to qubit 0
circ.ccx(0, 1, 2)      # Apply Toffoli gate controlled by qubits 0 and 1, target qubit 2
circ.h(0)              # Apply another Hadamard gate to qubit 0
circ.measure([0], [0]) # Measure qubit 0 into classical bit 0

# Generate LaTeX code using the latex drawer
fig = circ.draw(output='latex')
latex_source = circ.draw(output='latex_source')

# Save the LaTeX code to a .tex file
with open("circuit.tex", "w") as file:
    file.write(latex_source)

fig
```
