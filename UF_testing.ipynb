```python
from qiskit_qec.circuits import ArcCircuit
from qiskit_qec.decoders import UnionFindDecoder
from qiskit_aer import AerSimulator
from qiskit.providers.fake_provider import FakeSherbrooke
import numpy as np
```

```python
links = [[0, 1, 2], [2, 3, 4], [4, 5, 6], [6, 7, 8]]
T = 5 # number of syndrome measurement rounds
code = ArcCircuit(links, T, basis='zz',logical='1')
qc = code.circuit['zz']
#qc.draw(fold=-1)
```

```python
decoder = UnionFindDecoder(ArcCircuit(links, T))
counts = AerSimulator().run(qc).result().get_counts()
print(type(str(counts)))
print(str(counts))
```

```python
counts = FakeSherbrooke().run(qc, shots=1024).result().get_counts()

sorted_counts = sorted(counts.items(), key=lambda x: x[1], reverse=True)
print(sorted_counts)

```

```python
print(code.css_x_logical) # this gives the qubits on which the x logical operator acts
print(code.css_z_logical) # this gives the qubits on which the z logical operator acts (this is a physical choice)
measured_logicals = code.css_z_logical
net_z_logicals = {
                tuple(z_logical): 0 for z_logical in measured_logicals}
print(net_z_logicals)
```
